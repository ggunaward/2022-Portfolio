```python
#Library Imports
import numpy as np
import pandas as pd
import os.path
```

<p><u>Exercise 1</u></p>

<p><u>Exercise 2</u></p>

```python
#Check if csv exists and import data set into a variable
#Replace l's and 'n.a''s with sentinel values
if os.path.exists('Automobile_data.csv'):
    df2 = pd.read_csv('Automobile_data.csv', na_values={
        'company':["l","n.a"],
        'body-style':["l","n.a"],
        'wheel-base':["l","n.a"],
        'length':["l","n.a"],
        'engine-type':["l","n.a"],
        'num-of-cylinders':["l","n.a"],
        'horsepower':["l","n.a"],
        'average-mileage':["l","n.a"],
        'price':["l","n.a"]})
    
#Test for L's
    print(df2.iloc[13,5]=='l')
    print(pd.isnull(df2.iloc[13,5]))
    
#Update csv file
    df2.to_csv('Automobile_data.csv',index=False)
else:
    print ("File does not exist")
```

<p><u>Exercise 3</u></p>

```python
#Query dataframe
if 'df2' in locals():
    indexMaxPrice = list(np.where(df2['price'] == df2['price'].max())[0])
    df3=df2.loc[indexMaxPrice,['company','price']].copy()
    
#Display Results
    display(df3)
else:
    print ("File does not exist")
```

<p><u>Exercise 4</u></p>

```python
#query dataframe
if 'df2' in locals():
    indexToyota = list(np.where(df2['company'] == 'toyota')[0])
    df4 = df2.loc[indexToyota,:].copy()
    
#Display Results
    display(df4)
else:
    print ("File does not exist")
```

<p><u>Exercise 5</u></p>

```python
#Group by count of cars
if 'df2' in locals():
    df5 = pd.DataFrame(df2.groupby('company')[['company']].size(),columns=['']).copy()
    
#Display the results, sorted by count of cars descending
    print(df5.sort_values(by='',ascending=False))
else:
    print ("File does not exist")
```

<p><u>Exercise 6</u></p>

```python
#Group by company and aggregate data by highest priced car
if 'df2' in locals():
    indexCompanyPrice = ['company','price']
    df2GroupedCompany = pd.DataFrame(df2.groupby('company')[indexCompanyPrice].max()).copy()
    
#Display Results
    display(df2GroupedCompany)
else:
    print ("File does not exist")
```

<p><u>Exercise 7</u></p>

```python
#Group by company and aggregate data by Average of the Average mileage per car
if 'df2' in locals():
    df7 = df2.groupby(['company']).agg({'average-mileage': 'mean'}).copy()
    
#Display the Data
    display(df7)
else:
    print ("File does not exist")
```

<p><u>Exercise 8</u></p>

```python
#Sort the data and add to a new variable for display
if 'df2' in locals():
    df8=df2.sort_values(by='price',ascending=False)
    
#Display the Data
    pd.set_option('display.max_rows', None)
    display(df8.head())
else:
    print ("File does not exist")
```

<p><u>Exercise 9</u></p>

```python
#Obtain Dicts and assign to variables
germanCars = {'Company': ['Ford', 'Mercedes', 'BMV', 'Audi'], 'Price': [23845, 171995, 135925 , 71400]} 
japaneseCars = {'Company': ['Toyota', 'Honda', 'Nissan', 'Mitsubishi '], 'Price': [29995, 23600, 61500 , 58900]}

#Create DataFrames
dfGermanCars = pd.DataFrame(germanCars)
dfJapaneseCars = pd.DataFrame(japaneseCars)

#Concatenate DataFrames and create keys
df9 = pd.concat([dfGermanCars, dfJapaneseCars], axis=0, join="inner", keys=['Germany','Japan'])

#Display Results
display(df9)
```

<p><u>Exercise 10</u></p>

```python
#Obtain Dicts and assign to variables
Car_Price = {'Company': ['Toyota', 'Honda', 'BMV', 'Audi'], 'Price': [23845, 17995, 135925, 71400]} 
car_Horsepower = {'Company': ['Toyota', 'Honda', 'BMV', 'Audi'], 'horsepower': [141, 80, 182, 160]}

#Create Data Frames
dfCarPrice = pd.DataFrame(Car_Price)
dfCarHorsePower = pd.DataFrame(car_Horsepower)

#Merge the data frames with an inner join on Company
df10=pd.merge(dfCarPrice, dfCarHorsePower, on='Company')

#Display Data
display(df10)
```
