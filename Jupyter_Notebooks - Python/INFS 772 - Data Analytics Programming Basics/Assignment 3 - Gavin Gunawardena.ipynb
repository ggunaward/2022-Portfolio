```python
#Library Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Import data to a variable
df = pd.read_csv('company_sales_data.csv')

#Display imported data as dataframe
display(df)
```

<p><u>Exercise 1</u></p>

```python
#Configure figure and subplot
fig, axes = plt.subplots()
axes.set_yticks(ticks=[100000,200000,300000,400000,500000])
axes.set_ylabel('Profit per Month')
axes.set_xticks(range(1,13))
axes.set_xlabel('Month number')
axes.set_title('Company profit per month')
axes.set_ylim((100000.0, 500000.0))

#Add line graph to subplot
axes.plot(df[['month_number']],df[['total_profit']])

#Display
plt.show()
```

```python
#Configure figure and subplot
fig, axes = plt.subplots()
axes.set_yticks(ticks=[100000,200000,300000,400000,500000])
axes.set_ylabel('Sold units number')
axes.set_xticks(df['month_number'])
axes.set_xlabel('Month Number')
axes.set_title('Company Sales data of last year')
axes.set_ylim((100000.0, 500000.0))

#Add line graph to subplot
axes.plot(df[['month_number']],df[['total_profit']],
         linestyle = '--',
          linewidth=3.0,
          color='Red',
         marker = 'o',
         markerfacecolor='black',
         markeredgecolor='red',
         label='profit data of last year')

#Create and place legend
axes.legend(loc='lower right')

#Display
plt.show()
```

```python
#Configure figure and subplot
fig, axes = plt.subplots()
axes.set_yticks(ticks=[1000,2000,4000,6000,8000,10000,12000,15000,18000])
axes.set_ylabel('Sales units in number')
axes.set_xticks(df['month_number'])
axes.set_xlabel('Month Number')
axes.set_title('Sales data')
axes.set_ylim((0, 18000))

#Add line graphs to subplot
axes.plot(df[['month_number']],df['facecream'],
         color='blue',
         marker = 'o',
         markerfacecolor='blue',
         markeredgecolor='blue',
         label='Face cream Sales Data')
axes.plot(df[['month_number']],df['facewash'],
         color='orange',
         marker = 'o',
         markerfacecolor='orange',
         markeredgecolor='orange',
         label='Face Wash Sales Data')
axes.plot(df[['month_number']],df['toothpaste'],
         color='green',
         marker = 'o',
         markerfacecolor='green',
         markeredgecolor='green',
         label='ToothPaste Sales Data')
axes.plot(df[['month_number']],df['bathingsoap'],
         color='red',
         marker = 'o',
         markerfacecolor='red',
         markeredgecolor='red',
         label='Bathing Soap Sales Data')
axes.plot(df[['month_number']],df['shampoo'],
         color='purple',
         marker = 'o',
         markerfacecolor='purple',
         markeredgecolor='purple',
         label='Shampoo Sales Data')
axes.plot(df[['month_number']],df['moisturizer'],
         color='brown',
         marker = 'o',
         markerfacecolor='brown',
         markeredgecolor='brown',
         label='Moisturizer Sales Data')

#Create and place legend
axes.legend(loc='upper left')

#Display
plt.show()
```

```python
#Configure Subplot
fig, axes = plt.subplots()
axes.set_ylim((0.0, 31000.0))
axes.set_yticks(ticks=range(0,35000,5000))
axes.set_ylabel('Sold units in number')
axes.set_xticks(range(0,14,2))
axes.set_xlabel('Month Number')
axes.set_title('All product sales data using stack plot')

#Data Points
Months = df['month_number']
FaceCream = df['facecream']
FaceWash = df['facewash']
ToothPaste = df['toothpaste']
BathingSoap = df['bathingsoap']
Shampoo = df['shampoo']
Moisturizer = df['moisturizer']
Colors = ('purple','cyan','red','black','green','gold')
labels = ['Face Cream', 'Face Wash', 'Tooth Paste', 'Bathing Soap', 'Shampoo', 'Moisturizer']

#Add stack plot to subplot
axes.stackplot(Months, FaceCream, FaceWash, ToothPaste, BathingSoap, Shampoo, Moisturizer, colors= Colors, labels=labels)

#Create legend
axes.legend(loc='upper left')

#Display Result
plt.show()
```
